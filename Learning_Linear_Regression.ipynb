```python
# FUNCTION OF THE GRADIENT DESCENT

def gradientDescent(initial_b, initial_m, points, learning_rate, iterations):
    b, m = initial_b, initial_m
    
    for i in range(iterations):
        b, m = evaluateGradient(b, m, points, learning_rate)
        
    return b, m

def evaluateGradient(b, m, points, learning_rate):
    b_gradient = 0
    m_gradient = 0
    N = float(len(points))
    for point in points:
        b_gradient += -(point[1] - (b + (m * point[0])))
        m_gradient += -point[0] * (point[1] - (b + (m * point[0])))
    
    b_gradient *= 2/N
    m_gradient *= 2/N
    
    new_b = b - (learning_rate * b_gradient)
    new_m = m - (learning_rate * m_gradient)
    return new_b, new_m

def calculateError(b, m , points):
    errorValue = 0
    
    for point in points:
        # square of y - (b + mx)
        errorValue += (point[1] - ( b + (m * point[0]) ))**2
    
    return errorValue / float(len(points))
```

```python
# TRYING OUT SOME REGRESSION
# INITIALIZATION

import numpy as np
import plotly
from plotly import graph_objs as graph

plotly.offline.init_notebook_mode(connected=True)

# DATA
points = np.genfromtxt('data.csv', delimiter=',')

# HYPERPARAMETERS
learning_rate = 0.00001
iterations = 100

# linear reg formula : y = mx + b (slope formula)
initial_b = 0
initial_m = 0

finalB, finalM = gradientDescent(initial_b, initial_m, points, learning_rate, iterations)
print(f"Final B = {finalB}")
print(f"Final M = {finalM}")

x, y = [], []
for point in points:
    x.append(point[0])
    y.append(point[1])
minX, maxX = min(x), max(x)

trace = graph.Scatter(
    x = x,
    y = y,
    mode = 'markers'
)
traceLine = graph.Scatter(
    x = [minX, maxX],
    y = [(finalB + finalM * minX), (finalB + finalM * maxX)],
    mode = 'lines'
)

data = [trace, traceLine]

# plot using plotly
plotly.offline.iplot(data)
```
